# Opinion Mining Tools Benchmarking

In this notebook we display the results of our benchmarking of several opinion mining tools


## Sentiment 

```python
import pandas as pd
import glob

results_files = glob.glob("results/en-sentiment-*.csv")


results_dfs = []

for lang in ["en", "es", "it"]:
    results_files = glob.glob(f"results/{lang}-sentiment-*.csv")

    dfs = []
    for file in results_files:
        dfs.append(pd.read_csv(file))

    df = pd.concat(dfs, ignore_index=True)

    # Drop Macro Precision and Recall
    df = df.drop(columns=['Macro Precision', 'Macro Recall'])

    results = df.pivot(index="Dataset", columns="Model", values="Macro F1")

    results["lang"] = lang

    results_dfs.append(results.reset_index())


results = pd.concat(results_dfs)

results.set_index(["lang", "Dataset"], inplace=True)

res = (results * 100).round(1)[["pysentimiento", "tweetnlp", "stanza", "flair", "textblob", "vader"]]


res.fillna("-").astype(str)

```
